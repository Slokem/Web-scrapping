```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as rq
```

```python
import datetime
```

```python
import pandas as pd 
import numpy as np
```

```python
# First let's get the URL adress of the AMI international website we want to parse and have a database of all their events, title, address, description 
base_url = 'https://www.ami.international/events/upcoming' 
# Then according to the documentation of the library "Requests", the syntax 
# for how to "load" a webpage through Python is the following one 
r = rq.get(base_url)
# Then we strore in a BeautifulSOup Object the HTML text of the webpage 
soup = bs(r.text)
# the function prettify allow to display a Beautiful Object with tabulation like decent HTLM code
#print(soup.prettify())
```

```python
print(soup.title)
# it gives you the title of the hTLM code

print(soup.title.name)
#it giving the name of the element

print(soup.title.string)
# it gives the string content of the element

print(soup.title.parent.name)
# gives the name of the element above the previous one 

print(soup.p)

print(soup.a)
```

```python
print(' Time of extract is : ', datetime.datetime.today(),'\n ----- \n', 'criteria of research are : ', ", ", "\n ")
```

```python
#It seems like in the class name there will be always the mention "esl82me0", in that case we could use 
# the re library and make a research on the class_ content by using re.compile
import re
AMIsoup=soup.find_all("img", alt=True)
AMIsoup
TitleArray=[]
for Title in AMIsoup:
    #print(Title['alt'])
    TitleArray.append(Title['alt'])
TitleArray

#for HREF in AMIsoup:
#    print(HREF['href'].strip())
    
```

```python
AMIsoupaddress=soup.find_all("h4")
AMIsoupaddress
AddressArray=[]
for address in AMIsoupaddress:
    #print(address.string)
    AddressArray.append(address.string)
AddressArray
```

```python
AMIsoupdate=soup.find_all("h3")
AMIsoupdate
DateArray=[]
for Date in AMIsoupdate:
    DateArray.append(Date.string)
DateArray
```

```python
AMIsoupdesc=soup.find_all("em")
AMIsoupdesc
DescArray=[]
for Description in AMIsoupdesc:
    DescArray.append(Description.string)
len(DescArray)
DescArray
```

```python
#conveting all the array in Numpy arrays 

Title=np.array(TitleArray)
Address=np.array(AddressArray)
Date = np.array(DateArray)
Description = np.array(DescArray)
```

```python
# putting everything in a dataframe with pandas 

df=pd.DataFrame({"Title":Title,"Address" : Address, "Date": Date, "Description": Description})
df.head()
```
